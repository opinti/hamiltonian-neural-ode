```python
import numpy as np
from mechanical_systems import DoubleMassSpring, DoublePendulum
from plot_utils import plot_canonical_time_series, plot_phase_space_quiver
```

```python
double_mass_spring = DoubleMassSpring()

t_init = 0
t_final = 5
t_eval = np.linspace(t_init, t_final, 500)

data = double_mass_spring.simulate(
    q1 = -0.5,
    q2 = 1.,
    q1_dot = -0.5,
    q2_dot = 1,
    t_span=(t_init, t_final),
    t_eval=t_eval,
)

energy_hisory = data["energy"]
initial_energy = data["initial_energy"]

energy_mismatch = np.sqrt(np.mean(np.square(energy_hisory - initial_energy))) / initial_energy
assert energy_mismatch < 5e-2

plot_canonical_time_series(data["t"], data["canonical_coords"])

# double_mass_spring.create_gif(
#     data["t"],
#     data["cartesian_coords"],
#     interval=50,
# )

plot_phase_space_quiver(
    data["t"],
    data["generalized_coords"],
    quiver=False,
    trail=True,
)
```

```python
double_pendulum = DoublePendulum()

t_init = 0
t_final = 20
t_eval = np.linspace(t_init, t_final, 400)

data = double_pendulum.simulate(
    q1 = -2,
    q2 = -0.5,
    q1_dot = -0.2,
    q2_dot = -0.1,
    t_span=(t_init, t_final),
    t_eval=t_eval,
)

energy_hisory = data["energy"]
initial_energy = data["initial_energy"]

energy_mismatch = np.sqrt(np.mean(np.square(energy_hisory - initial_energy))) / initial_energy
assert energy_mismatch < 5e-2

plot_canonical_time_series(data["t"], data["canonical_coords"])

double_pendulum.create_gif(
    data["t"],
    data["cartesian_coords"],
    interval=50,
    trail_length=20,
)

plot_phase_space_quiver(
    data["t"],
    data["generalized_coords"],
    quiver=False,
)
```
